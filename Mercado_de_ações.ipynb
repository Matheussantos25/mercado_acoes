```python
import yfinance as yf

# Definindo o ticker para ibovespa
ibovespa = yf.Ticker("^BVSP")

# Baixando o histórico de preços
df = ibovespa.history(period="1d", start="2013-01-01", end="2023-01-01")

# Resetando o índice do DataFrame e selecionando apenas as colunas 'Date' e 'Close'
df_ibov = df.reset_index()[['Date', 'Close']]

# Extraia o ano da coluna 'Date' para criar uma nova coluna 'Year'
df_ibov['Year'] = df_ibov['Date'].dt.year

# Agrupe o dataframe por 'Year' e calcule o desvio padrão para a coluna 'Close' em cada grupo
std_dev_per_year_ibovespa = df_ibov.groupby('Year')['Close'].std()

std_dev_per_year_ibovespa

```

```python
import yfinance as yf

# Definindo o ticker para ibovespa
ibovespa = yf.Ticker("^BVSP")

# Baixando o histórico de preços para janeiro de 2023
df_ibov = ibovespa.history(period="1d", start="2023-01-01", end="2023-01-31")

# Resetando o índice do DataFrame e selecionando apenas as colunas 'Date' e 'Close'
df_ibov = df_ibov.reset_index()[['Date', 'Close']]

# Extraia a semana do ano da coluna 'Date' para criar uma nova coluna 'Week'
df_ibov['Week'] = df_ibov['Date'].dt.isocalendar().week

# Agrupe o dataframe por 'Week' e calcule o desvio padrão para a coluna 'Close' em cada grupo
std_dev_per_week_jan2023 = df_ibov.groupby('Week')['Close'].std()

std_dev_per_week_jan2023
```

```python
import yfinance as yf
import pandas as pd

# Lista de tickers
tickers = ["ABEV3.SA", "BBDC3.SA", "BBDC4.SA", "EMBR3.SA", "GGBR3.SA", "GGBR4.SA", "ITSA3.SA", "ITSA4.SA", "ITUB3.SA", "ITUB4.SA", "PETR3.SA", "PETR4.SA", "SANB3.SA", "SANB4.SA", "SANB11.SA", "USIM3.SA", "USIM5.SA", "USIM6.SA", "VALE3.SA"]

# Inicializar um DataFrame vazio para armazenar os resultados
df_final = pd.DataFrame()

# Baixar os preços de fechamento para cada ticker
for ticker in tickers:
    data = yf.download(ticker, start="2013-01-01", end="2023-01-01")
    data = data.reset_index()[['Date', 'Close']]
    data.columns = ['Date', ticker]

    # Mesclar os resultados em df_final
    if df_final.empty:
        df_final = data
    else:
        df_final = pd.merge(df_final, data, on='Date', how='outer')

df_final
```

```python
# Baixando todos os dados de uma vez
data_all = yf.download(tickers, start="2013-01-01", end="2013-12-31")['Close']

# Calculando a valorização percentual para cada ticker
valorizacao = ((data_all.iloc[-1] - data_all.iloc[0]) / data_all.iloc[0]) * 100

# Ordenando os tickers pela valorização percentual em ordem decrescente
valorizacao_sorted = valorizacao.sort_values(ascending=False)

df_valorizacao = valorizacao_sorted.reset_index()
df_valorizacao.columns = ['Ticker', 'Valorização (%)']

df_valorizacao
```
